```python
import numpy as np
import pandas as pd
import collections
```

```python
#please put the files within a same direction
q=pd.read_csv('Questions.csv', encoding='latin-1')
t=pd.read_csv('Tags.csv')
qnum=q.shape[0]
#print(qnum)
#t.head(3)
```

```python
#appending tags of a same question into one row
tappend= t['Tag'].groupby([t.Id]).apply(list).reset_index()
tappend.head(3)
```

```python
n=10
tagCount10 =  collections.Counter(list(t['Tag'])).most_common(n)
print(tagCount10)
tagCount10np=np.array(tagCount10)
tagCount10np.reshape(n,2)
tagCount10pd=pd.DataFrame(tagCount10np)
tagCount10pd.columns=['tag', 'count']
tagCount10pd['count']=(tagCount10pd['count']).astype(int)
tagCat10=list(tagCount10pd['tag'])
```

```python
a10 = np.zeros(shape=(qnum,n))
iter=-1
for d in tappend.Tag:
    iter=iter+1
    for k in range(n):
        if tagCat10[k] in d:
            a10[iter, k]=1
```

```python
a10=pd.DataFrame(a10, columns=tagCat10)
at10=pd.concat([tappend, a10], axis=1)
col_list= list(at10)
col_list.remove('Id')
at10['sumtag'] = at10[col_list].sum(axis=1)
at10.head(3)
#at10.to_csv('tagappendforquestion.csv', encoding='utf-8', index=False)
```
